#Data Visualization
-------------------

This notebook is about different ways data visualization transforms data into images, making complex information accessible and understandable. Hopefully, at the most basic level, the image should be able to tell a story that as many people as possible can look at the image clearly identify what is happening.

#Bar Chart
----------

Bar Charts are a simple, frequently used, and effective ways to read the story of data. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample DataFrame
sales_summary = pd.DataFrame({
    'Product': ['Battery', 'Inverter', 'Solar Panel'],
    'Total_Sales': [4200, 2500, 4250]
})

plt.figure(figsize=(8,6))
sns.barplot(x='Product', y='Total_Sales', data=sales_summary, palette='viridis')
plt.title('Total Sales per Product')
plt.xlabel('Product')
plt.ylabel('Total Sales ($)')
plt.show()
```

#Line Plots
-----------

Line Plots are another frequently used data visualization tool, specifically they can be great for time series data. 

```python
# Sample DataFrame
data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Sales': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550]
}
df_line = pd.DataFrame(data)

plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Sales', data=df_line, marker='o')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#Scatter Plot
-------------

Scatter Plots can often be a good place to start when first inspecting data, as it can give a good sense of the relationship between your two varaibles and how distributed your data may be.

```python
# Sample DataFrame
data = {
    'Marketing_Spend': [1000,1500, 2000, 2500, 3000, 3500, 4000],
    'Sales': [200, 300, 400, 500, 600, 700, 800]
}
df_scatter = pd.DataFrame(data)

plt.figure(figsize=(8,6))
sns.scatterplot(x='Marketing_Spend', y='Sales', data=df_scatter, s=100, color='green')
plt.title('Sales vs. Marketing Spend')
plt.xlabel('Marketing Spend ($)')
plt.ylabel('Sales Volume')
plt.show()
```

#Heatmaps
---------

Heatmaps can be good to show the correlations between multiple variables as well as the density. It was quite helpful for me when looking at the output of the Agent-Based Model(ABM) I created concerning social dynamics as well as the ABM concerning the interactions between human and generative ai, which you can find in my repository. 

```python
# Sample DataFrame
data = {
    'Sales': [100, 150, 200, 300, 300],
    'Marketing_Spend': [1000, 1000, 2000, 2500, 3000],
    'Economic_Indicator': [80, 85, 90, 90, 100]
}
df_heatmap = pd.DataFrame(data)

plt.figure(figsize=(8,6))
corr_matrix = df_heatmap.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
